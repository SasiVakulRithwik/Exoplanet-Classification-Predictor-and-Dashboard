```python
# %% [markdown]
# # Exoplanet Dataset - Exploratory Data Analysis
# 
# This notebook performs exploratory data analysis on the NASA Exoplanet dataset to understand the data structure, distributions, and relationships before building our machine learning model.

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")

# %%
# Load the dataset
data_path = Path('../data/Exoplanet-Dataset.csv')
df = pd.read_csv(data_path)

print("Dataset loaded successfully!")
print(f"Dataset shape: {df.shape}")

# %%
# Basic dataset information
print("=== DATASET BASIC INFO ===")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
print("\nColumn names:")
print(df.columns.tolist())

# %%
# Display first few rows
print("=== FIRST 5 ROWS ===")
display(df.head())

# %%
# Display last few rows
print("=== LAST 5 ROWS ===")
display(df.tail())

# %%
# Data types and basic info
print("=== DATASET INFO ===")
df.info()

# %%
# Check for missing values
print("=== MISSING VALUES ANALYSIS ===")
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100

missing_summary = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing Percentage': missing_percent
})

# Filter to show only columns with missing values
missing_summary = missing_summary[missing_summary['Missing Count'] > 0]

if missing_summary.empty:
    print("🎉 No missing values found in the dataset!")
else:
    print("Columns with missing values:")
    print(missing_summary)

# %%
# Basic statistical summary
print("=== STATISTICAL SUMMARY ===")
display(df.describe())

# %%
# Analyze the target variable 'label'
print("=== TARGET VARIABLE (label) ANALYSIS ===")

# First, let's see what unique values we have in the label column
unique_labels = df['label'].unique()
print(f"Unique label values: {sorted(unique_labels)}")

# Create dynamic label mapping based on actual values in dataset
label_names = {}
for label_val in unique_labels:
    if label_val == 0:
        label_names[label_val] = 'False Positive'
    elif label_val == 1:
        label_names[label_val] = 'Candidate'
    elif label_val == 2:
        label_names[label_val] = 'Confirmed Planet'
    else:
        label_names[label_val] = f'Unknown Class {label_val}'

print("Label mapping:")
for key, value in label_names.items():
    print(f"  {key}: {value}")

# Value counts and percentages
label_counts = df['label'].value_counts().sort_index()
label_percentages = (df['label'].value_counts(normalize=True) * 100).sort_index()

label_summary = pd.DataFrame({
    'Count': label_counts,
    'Percentage': label_percentages
})
label_summary['Class Name'] = label_summary.index.map(label_names)

print("\nLabel distribution summary:")
display(label_summary)

# %%
# Visualize label distribution
plt.figure(figsize=(12, 5))

# Bar plot
plt.subplot(1, 2, 1)
class_names = [label_names[i] for i in label_summary.index]
bars = plt.bar(class_names, label_summary['Count'], 
              color=['#ff6b6b', '#ffe66d', '#4ecdc4', '#95e1d3', '#f38181'][:len(class_names)], 
              alpha=0.8)
plt.title('Distribution of Exoplanet Classes', fontsize=14, fontweight='bold')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Add count labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom')

# Pie chart
plt.subplot(1, 2, 2)
colors = ['#ff6b6b', '#ffe66d', '#4ecdc4', '#95e1d3', '#f38181'][:len(class_names)]
plt.pie(label_summary['Count'], labels=class_names, 
        autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Class Distribution (%)')

plt.tight_layout()
plt.show()

# %%
# Distribution of numerical features
print("=== NUMERICAL FEATURES DISTRIBUTION ===")

# Select numerical columns (exclude label for now)
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# Remove 'label' from numerical features for feature analysis
feature_cols = [col for col in numerical_cols if col != 'label']

print(f"Number of numerical features: {len(feature_cols)}")
print("Numerical features:", feature_cols)

# %%
# Plot distributions of key features
key_features = ['PC1', 'PC2', 'PC3', 'planet_radius', 'orbital_period', 
                'star_teff', 'star_logg', 'star_radius', 'star_mass']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for i, feature in enumerate(key_features):
    if i < len(axes):
        # Check if feature exists in dataframe
        if feature in df.columns:
            # Histogram with KDE
            df[feature].hist(bins=50, ax=axes[i], alpha=0.7, color='skyblue', edgecolor='black')
            axes[i].set_title(f'Distribution of {feature}', fontsize=12)
            axes[i].set_xlabel(feature)
            axes[i].set_ylabel('Frequency')
            
            # Add vertical line for mean
            mean_val = df[feature].mean()
            axes[i].axvline(mean_val, color='red', linestyle='--', linewidth=1, label=f'Mean: {mean_val:.2f}')
            axes[i].legend()
        else:
            axes[i].set_title(f'{feature} not found', fontsize=12)
            axes[i].text(0.5, 0.5, 'Feature not found', ha='center', va='center', transform=axes[i].transAxes)

# Remove empty subplots
for i in range(len(key_features), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# %%
# Box plots for key features to identify outliers
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

plotted_features = 0
for i, feature in enumerate(key_features):
    if i < len(axes) and feature in df.columns:
        df.boxplot(column=feature, ax=axes[i], grid=False)
        axes[i].set_title(f'Box Plot of {feature}', fontsize=12)
        plotted_features += 1

# Remove empty subplots
for i in range(plotted_features, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# %%
# Correlation analysis
print("=== CORRELATION ANALYSIS ===")

# Calculate correlation matrix with only existing numerical columns
available_features = [col for col in feature_cols if col in df.columns]
correlation_matrix = df[available_features + ['label']].corr()

# Plot correlation heatmap
plt.figure(figsize=(16, 14))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8}, fmt='.2f')
plt.title('Feature Correlation Matrix', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# %%
# Top correlations with target variable
print("=== TOP CORRELATIONS WITH TARGET (label) ===")
label_correlations = correlation_matrix['label'].sort_values(ascending=False)

# Remove label correlation with itself
label_correlations = label_correlations[label_correlations.index != 'label']

print("Top 10 features most correlated with label:")
top_correlations = label_correlations.head(10)
for feature, corr in top_correlations.items():
    print(f"{feature}: {corr:.4f}")

# %%
# Feature distributions by class
print("=== FEATURE DISTRIBUTIONS BY CLASS ===")

# Select top 4 features correlated with label for visualization
top_features = label_correlations.head(4).index.tolist()

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

colors = ['#ff6b6b', '#ffe66d', '#4ecdc4', '#95e1d3', '#f38181']

for i, feature in enumerate(top_features):
    if i < len(axes):
        for j, label_value in enumerate(label_summary.index):
            if j < len(colors):  # Ensure we don't exceed color list
                subset = df[df['label'] == label_value]
                if len(subset) > 0:
                    axes[i].hist(subset[feature], bins=30, alpha=0.6, color=colors[j], 
                                label=label_names[label_value], density=True)
        
        axes[i].set_title(f'Distribution of {feature} by Class', fontsize=12)
        axes[i].set_xlabel(feature)
        axes[i].set_ylabel('Density')
        axes[i].legend()

# Remove empty subplots
for i in range(len(top_features), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# %%
# Outlier analysis using IQR method
print("=== OUTLIER ANALYSIS ===")

def detect_outliers_iqr(feature):
    Q1 = feature.quantile(0.25)
    Q3 = feature.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = feature[(feature < lower_bound) | (feature > upper_bound)]
    return len(outliers), lower_bound, upper_bound

outlier_summary = []
for feature in feature_cols:
    if feature in df.columns:
        outlier_count, lower, upper = detect_outliers_iqr(df[feature])
        outlier_percent = (outlier_count / len(df)) * 100
        outlier_summary.append({
            'Feature': feature,
            'Outlier Count': outlier_count,
            'Outlier Percentage': outlier_percent,
            'Lower Bound': lower,
            'Upper Bound': upper
        })

outlier_df = pd.DataFrame(outlier_summary)
print("Outlier summary (showing features with outliers):")
outliers_present = outlier_df[outlier_df['Outlier Count'] > 0]
if len(outliers_present) > 0:
    display(outliers_present.sort_values('Outlier Percentage', ascending=False))
else:
    print("No outliers detected using IQR method.")

# %%
# Feature scales analysis
print("=== FEATURE SCALES ANALYSIS ===")

scale_analysis = []
for col in feature_cols:
    if col in df.columns:
        scale_analysis.append({
            'Feature': col,
            'Min': df[col].min(),
            'Max': df[col].max(),
            'Mean': df[col].mean(),
            'Std': df[col].std(),
            'Range': df[col].max() - df[col].min()
        })

scale_df = pd.DataFrame(scale_analysis)
print("Feature scales summary (top 10 by range):")
display(scale_df.sort_values('Range', ascending=False).head(10))

# %%
# Summary of key findings
print("="*50)
print("📊 EDA SUMMARY")
print("="*50)

print(f"\n📈 DATASET OVERVIEW:")
print(f"   • Total records: {df.shape[0]}")
print(f"   • Total features: {df.shape[1]}")
print(f"   • Missing values: {'None' if df.isnull().sum().sum() == 0 else df.isnull().sum().sum()}")

print(f"\n🎯 TARGET DISTRIBUTION:")
for label_val in sorted(unique_labels):
    count = len(df[df['label'] == label_val])
    percentage = (count / len(df)) * 100
    print(f"   • {label_names[label_val]}: {count} ({percentage:.1f}%)")

print(f"\n🔍 KEY INSIGHTS:")
if len(label_correlations) > 0:
    print(f"   • Most correlated feature with target: {label_correlations.index[0]} ({label_correlations.iloc[0]:.3f})")
    print(f"   • Least correlated feature with target: {label_correlations.index[-1]} ({label_correlations.iloc[-1]:.3f})")

# Check for class imbalance
max_class_pct = max(label_percentages)
if max_class_pct > 60:
    print(f"   ⚠️  Potential class imbalance: Largest class is {max_class_pct:.1f}% of data")

print(f"\n📏 FEATURE SCALES:")
if len(scale_df) > 0:
    max_range_feature = scale_df.loc[scale_df['Range'].idxmax(), 'Feature']
    max_range = scale_df.loc[scale_df['Range'].idxmax(), 'Range']
    print(f"   • Feature with largest range: {max_range_feature} (range: {max_range:.2f})")

print(f"\n⚡ RECOMMENDATIONS FOR MODELING:")
print(f"   • Feature scaling will be necessary (features have different scales)")
if len(outliers_present) > 0:
    max_outlier_feature = outliers_present.loc[outliers_present['Outlier Percentage'].idxmax(), 'Feature']
    print(f"   • Consider handling outliers in features like {max_outlier_feature}")
if df.isnull().sum().sum() > 0:
    print(f"   • Need to handle missing values before training")
print(f"   • Random Forest should handle correlations well")

# %%
# Check for any infinite values
print("=== CHECKING FOR INFINITE VALUES ===")
infinite_counts = np.isinf(df.select_dtypes(include=[np.number])).sum()
infinite_features = infinite_counts[infinite_counts > 0]
if infinite_features.empty:
    print("✅ No infinite values found!")
else:
    print("Infinite values found:")
    print(infinite_features)

# %%
# Check for constant features
print("=== CHECKING FOR CONSTANT FEATURES ===")
constant_features = []
for col in feature_cols:
    if col in df.columns and df[col].nunique() <= 1:
        constant_features.append(col)

if constant_features:
    print(f"Constant features found: {constant_features}")
else:
    print("✅ No constant features found!")

print("\n✅ EDA Complete! Ready for model training.")
```
